## Scatterplot

## Stack Plots

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
passear = [8,5,7,8,13]
```

```python
plt.stackplot(dias, dormir, comer, trabalhar, passear, colors= ['m','c','r','k','b'])
plt.show()
```

## Pie Chart

```python
fatias = [7,2,2,13]
atividades = ['dormir', 'comer', 'trabalhar', 'passear']
colunas = ['c', 'm', 'r', 'k']
```

```python
plt.pie(fatias, labels= atividades, colors= colunas, startangle= 90, shadow= True, explode = (0,0.1,0,0))
plt.show()
```

##  Pylab

```python
# Visualizando os gráficos dentro do Jupyter Notebook
# O Pylab combina funcionalidades do pyplot com funcionalidade do Numpy
from pylab import *
%matplotlib inline
```

```python
x = linspace(0, 5, 10)
#Linspace retorna números espaçados de modo uniforme em um intervalo. Dessa forma, dado um ponto inicial e de parada,
#assim como a quantidade de valores, linspace irá distribuí-los uniformemente para você em uma matriz NumPy
y = x ** 2

fig = plt.figure()

# Definindo os eixos
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')
```

```python
# Gŕaficos com 2 figuras
x = linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # eixos da figura principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # eixos da figura secundária

#Figura principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Figura Principal')


#Figura Secundária
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundária')
```

```python
# Gráficos em Paralelo
fig, axes = plt.subplots(nrows= 1, ncols= 2)

for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')
    
fig.tight_layout()    
```
