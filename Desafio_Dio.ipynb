# O rodízio municipal de veículos

## Forma 01

```python
def maiuscula(p):
    if any(i.isupper() for i in p):
        return True
    return False

n = int(input())
while n:
    n -= 1
    placa = input().split('-')
    qtdDeItens = len(placa)
    if (qtdDeItens == 2):
        numero = placa.pop()
        numero = numero[-1]
        letras = placa[0]
        if ( maiuscula(letras) == True and len(letras) == 3 ):
            if (int(numero) in (1,2)):
                print('Segunda-Feira')
            elif (int(numero) in (3,4)):
                print('Terça-Feira')
            elif (int(numero) in (5,6)):
                print('Quarta-Feira')
            elif (int(numero) in (7,8)):
                print('Quinta-Feira') 
            elif (int(numero) in (9,0)):
                print('Sexta-Feira')          
            else:
                print('Falha')
        else:
                print('Falha')
    else:
        print('Falha')       
```

# A mudança

```python
while True:

  try:



    graus = int(input())

     

    if graus == 360 or graus >= 0 and graus < 90:

     msg = 'Bom Dia!!'

    elif graus >= 90 and graus < 180:

     msg = 'Boa Tarde!!'

    elif graus >= 180 and graus < 270:

     msg = 'Boa Noite!!'

    else :

     msg = 'De Madrugada!!'

     

    print(msg)  





  except EOFError:

    break
```
