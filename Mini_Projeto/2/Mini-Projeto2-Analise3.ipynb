```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
list(df.columns.values)
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio - solução DSA 
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#00e600", "#ff8c1a","#a180cc"]
sns.barplot(x="fuelType", y="price",hue="gearbox", palette="husl",data=df)
ax.set_title("Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio",fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de Combustível",fontdict= {'size':14})
ax.yaxis.set_label_text("Preço Médio",fontdict= {'size':14})
plt.show()
```

```python
# minha solução

## preço = price
## tipo de  de Combustível = fuelType
## tipo de caixa de câmbio = gearbox

fig, ax = plt.subplots(figsize=(8,5))
colors = ["#EA899A", "#77DD77","#4682B4"]
sns.barplot(x="fuelType", y="price",hue="gearbox", palette=colors, data=df)
ax.set_title("Preço médio dos veículos por tipo de combustível e tipo de caixa de câmbio",fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de Combustível",fontdict= {'size':12})
ax.yaxis.set_label_text("Preço Médio",fontdict= {'size':12})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/fueltype_vehicleType_price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio -solução da dsa 

colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"]
fig, ax = plt.subplots(figsize=(8,5))
sns.set_palette(sns.xkcd_palette(colors))
sns.barplot(x="vehicleType", y="powerPS",hue="gearbox",data=df)
ax.set_title("Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio",fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':14})
ax.yaxis.set_label_text("Potência Média",fontdict= {'size':14})
plt.show()
```

```python
# minha solução

## potencia = powerPS
## tipo de veículo = vehicleType
## tipo de caixa de câmbio = gearbox

fig, ax = plt.subplots(figsize=(8,5))
colors = ["#EA899A", "#77DD77","#4682B4"]
sns.barplot(x="vehicleType", y="powerPS",hue="gearbox", palette=colors, data=df)
ax.set_title("Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio",fontdict= {'size':12})
ax.xaxis.set_label_text("Tipo de Veículo",fontdict= {'size':12})
ax.yaxis.set_label_text("Preço Médio",fontdict= {'size':12})
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise3/vehicletype_fueltype_power.png")
```
